```python
import pandas as pd 
```

# Mục mới

```python
# a)
apl = pd.read_csv("/content/aapl.csv")
amzn = pd.read_csv("/content/amzn.csv")
fb = pd.read_csv("/content/fb.csv")
goog = pd.read_csv("/content/goog.csv")
nflx = pd.read_csv("/content/nflx.csv")
```

```python
def add_column(data,name:str):
  return data.assign(ticker = name)
```

```python
apl = add_column(apl,'APPl')
amzn = add_column(amzn,'AMZ')
fb = add_column(fb,'FB')
goog = add_column(goog,'GOG')
nflx = add_column(nflx,'NFL')
```

```python
# d)
faang = pd.concat([goog,apl,fb,amzn,nflx])
faang
```

```python
faang = faang.assign(date = pd.to_datetime(faang.date),
                     volume_2 = faang.volume.astype(int))
faang
```

```python
# f: Find the seven rows in faang with the lowest value for volume
D = faang.sort_values(by = ['volume_2'],axis = 0 , ascending= True).head(7)
print(D)
```

```python
# Ex2:a) 
covid = pd.read_csv("/content/covid19_cases.csv")
```

```python
# b) Create a date column using the data 
# in the dateRep column and the pd.to_datetime() function. 
covid = covid.assign(date = pd.to_datetime(covid.dateRep))
covid.head()
```

```python
# c) Set the date column as the index and sort the index.
covid = covid.sort_values(by = ['date'],ascending=True)
covid = covid.set_index('date')
covid
```

```python
covid.rename(columns= {'countriesAndTerritories': 'CTT'}, inplace = True)
covid.columns
```

```python
covid['CTT'] = covid.CTT.str.replace('United_States_of_America','USA')
```

```python
covid['CTT'] = covid.CTT.str.replace('United_Kingdom','UK')
```

```python
B = covid[covid.CTT.str.contains('Argentina|Brazil|China|Colombia|India|Italy|Mexico|Peru|Russia|Spain|Turkey|UK|USA')]
H = B.groupby(['CTT']).sum()
print(H[['cases']])
```

```python
# Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the
# case counts (the cases column). Be sure to fill in NaN values with 0.
covid['case'] = covid['cases'].fillna(0)
pivot_cov = covid.pivot(columns='CTT',values='cases')
print(pivot_cov.head)
```
